```python
combined_input = np.zeros((35001,9))
combined_output = np.zeros((35001,7))
```

```python
with open('data/Compute_CSV_input.csv', 'r') as fo:
    linecount=0
    reader = csv.reader(fo)
    for row in reader:
        for column in range(9):
            combined_input[linecount][column]=row[column]
        linecount=linecount+1
        if linecount > 35000:
            break
```

```python
with open('data/Compute_CSV_output.csv', 'r') as fo:
    linecount=0
    reader = csv.reader(fo)
    for row in reader:
        for column in range(7):
            combined_output[linecount][column]=row[column]
        linecount=linecount+1
        if linecount > 35000:
            break
```
